```python
m=df['horsepower'].median()
```

```python
m
```

```python
low_horsepower=df.query('horsepower<{}'.format(m))
```

```python
high_horsepower=df.query('horsepower>{}'.format(m))
```

```python
high_horsepower
```

```python
low_horsepower
```

```python
mean_horsepower_of_low_horsepower=low_horsepower['horsepower'].mean()
```

```python
mean_horsepower_of_high_horsepower=high_horsepower['horsepower'].mean()
```

```python
location=[1,2]
heights=[mean_horsepower_of_low_horsepower,mean_horsepower_of_high_horsepower]
labels=["low","high"]
plt.bar(location,heights,tick_label=labels)
plt.title("average horsepower")
plt.xlabel('horsepower')
plt.ylabel('average horsepower')
plt.show()
```
